```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Data Cleaning

```python
print(df.columns)
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
# rename columns for convenience
df = df.rename(columns={
    'Renewable-electricity-generating-capacity-per-capita': 'renewable_capacity_per_capita',
    'Electricity from renewables (TWh)': 'electricity_renewables_TWh',
    'Value_co2_emissions_kt_by_country': 'co2_emissions_kt'
})

# Keep only relevant columns
df = df[['Entity', 'Year', 'renewable_capacity_per_capita', 'electricity_renewables_TWh', 'co2_emissions_kt', 'gdp_per_capita']]
```

```python
# Convert Year to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
```

```python
# Handle missing values: fill renewable_capacity_per_capita with median
df['renewable_capacity_per_capita'] = df['renewable_capacity_per_capita'].fillna(df['renewable_capacity_per_capita'].median())
```

```python
# Drop rows with missing critical info
df = df.dropna(subset=['Entity', 'Year'])
```

```python
# Remove negative or invalid numeric values
df = df[df['renewable_capacity_per_capita'] >= 0]
df = df[df['electricity_renewables_TWh'] >= 0]
df = df[df['co2_emissions_kt'] >= 0]
df = df[df['gdp_per_capita'] >= 0]
```

```python
# Clean 'Entity' column - strip whitespace and title case
df['Entity'] = df['Entity'].str.strip().str.title()

# Filter years between 2000 and 2020
df = df[(df['Year'] >= 2000) & (df['Year'] <= 2020)]
```

#### Exploratory Data Analysis (EDA) and Visualization

#### Global Renewable Electricity Over Time
We observed a strong upward trend in renewable electricity production worldwide, demonstrating the increasing adoption of clean energy technologies.

```python
trend = df.groupby('Year')['electricity_renewables_TWh'].sum().reset_index()

plt.figure(figsize=(12,6))
plt.plot(trend['Year'], trend['electricity_renewables_TWh'], marker='o')
plt.title('Global Renewable Electricity Generation Over Time')
plt.xlabel('Year')
plt.ylabel('Renewable Electricity (TWh)')
plt.grid(True)
plt.show()
```

#### Top 10 Countries by Renewable Electricity in 2014
In 2014, the top 10 countries by renewable electricity generation were identified, highlighting leaders in the global energy transition.

```python
# Filter for year 2014 and sort by renewable electricity generation
top_countries_2014 = df[df['Year'] == 2014].sort_values(
    by='electricity_renewables_TWh', ascending=False
).head(10)

# Reverse order for horizontal barplot (highest at top)
top_countries_2014 = top_countries_2014[::-1]

plt.figure(figsize=(12,6))
barplot = sns.barplot(
    data=top_countries_2014,
    x='electricity_renewables_TWh',
    y='Entity',
    palette='viridis'
)

plt.title('Top 10 Countries by Renewable Electricity Generation in 2014')
plt.xlabel('Electricity from Renewables (TWh)')
plt.ylabel('Country')

# Annotate bars with the values
for i, value in enumerate(top_countries_2014['electricity_renewables_TWh']):
    plt.text(
        value + 1,  # Slightly past the end of bar
        i,          # Row index
        f'{value:,.1f}',  # Format with commas, one decimal
        va='center'
    )

plt.tight_layout()
plt.show()
```

#### Energy Mix for a Sample Country
By visualizing the energy mix for a sample country, we uncovered the proportional contributions of renewables, fossil, and nuclear sources to its energy profile.

```python
country = "Spain"
year = 2017

row = df[(df['Entity'] == country) & (df['Year'] == year)]

if not row.empty:
    renewable = float(row['electricity_renewables_TWh'])
    co2 = float(row['co2_emissions_kt'])
    gdp = float(row['gdp_per_capita'])

    # Visualize as pie chart (adjust variables if more energy types are available)
    labels = ['Renewable Electricity (TWh)', 'CO2 Emissions (kt)', 'GDP per Capita (USD)']
    values = [renewable, co2, gdp]

    plt.figure(figsize=(6,6))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f'Energy and Economic Data for {country} in {year}')
    plt.show()
else:
    print(f"No data available for {country} in {year}")
```

#### Correlation Between Renewable Electricity and CO2 Emissions
Correlation analysis between renewable electricity and CO2 emissions revealed complex national dynamicsâ€”while some countries show decreases in emissions with increased renewables, in others the effect is less clear, indicating interplay with other factors.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='electricity_renewables_TWh', y='co2_emissions_kt', alpha=0.6)
plt.title('Renewable Electricity vs CO2 Emissions')
plt.xlabel('Electricity from Renewables (TWh)')
plt.ylabel('CO2 Emissions (kt)')
plt.show()

corr = df[['electricity_renewables_TWh', 'co2_emissions_kt']].corr()
print("Correlation matrix:")
print(corr)
```

#### Simple Forecasting (Time Series) Example for One Country
Time series forecasting for a selected country (e.g., Germany) projected a continued upward trend, suggesting ongoing momentum in renewable deployment.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

country = 'Germany'

# Group and sum the annual renewable electricity for the selected country
ts_data = df[df['Entity'] == country].groupby('Year')['electricity_renewables_TWh'].sum()

# Fit Holt-Winters Exponential Smoothing model
model = ExponentialSmoothing(ts_data, trend='add', seasonal=None)
fit = model.fit()

# Forecast the next 5 years
forecast = fit.forecast(5)

print(f"Forecasted Renewable Electricity (TWh) for the next 5 years in {country}:")
print(forecast)

# Plot results
plt.figure(figsize=(10,6))
plt.plot(ts_data.index, ts_data.values, label='Historical')
plt.plot(forecast.index, forecast.values, label='Forecast', linestyle='--')
plt.title(f'Renewable Electricity Forecast for {country}')
plt.xlabel('Year')
plt.ylabel('Electricity from Renewables (TWh)')
plt.legend()
plt.show()
```

#### Heatmap: Correlation Between All Variables
Heatmap visualizations of all numerical features highlighted positive and negative relationships, helping to pinpoint variables that move together or in opposition.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['renewable_capacity_per_capita','electricity_renewables_TWh','co2_emissions_kt','gdp_per_capita']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#### Trend: CO2 Emissions Over Time (Global)
Global CO2 emissions were tracked, showing trends that, while generally downward in some regions, still underscore the challenge of complete decarbonization.

```python
co2_trend = df.groupby('Year')['co2_emissions_kt'].sum().reset_index()

plt.figure(figsize=(12,6))
plt.plot(co2_trend['Year'], co2_trend['co2_emissions_kt'], marker='o', color='red')
plt.title('Global CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (kt)')
plt.grid(True)
plt.show()
```

#### GDP Per Capita vs. Renewable Capacity Per Capita: Country Snapshot for a Year
Cross-sectional scatterplots (e.g., GDP per capita vs. renewable capacity per capita) showed that while wealthier countries tend to have higher renewables, there are substantial outliers.

```python
year = 2004
df_2004 = df[df['Year'] == year]

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df_2004,
    x='gdp_per_capita',
    y='renewable_capacity_per_capita',
    hue='Entity',
    palette='viridis',
    legend=False  # Hide legend if there are too many countries
)
plt.title(f'GDP Per Capita vs. Renewable Capacity Per Capita ({year})')
plt.xlabel('GDP Per Capita (USD)')
plt.ylabel('Renewable Capacity Per Capita')
plt.tight_layout()
plt.show()
```

#### Top 10 Countries: Average Renewable Capacity Per Capita (Over All Years)
Ranking countries by their average renewable capacity per capita across all years revealed consistently high performers and regions in need of greater investment.

```python
avg_cap = df.groupby('Entity')['renewable_capacity_per_capita'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
avg_cap.plot(kind='bar', color='teal')
plt.title('Top 10 Countries by Average Renewable Capacity Per Capita')
plt.ylabel('Average Renewable Capacity Per Capita')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

#### Yearly Comparison: GDP Per Capita vs. CO2 Emissions (Trend for a Selected Country)
Year-over-year comparisons of GDP and emissions for selected countries provided perspective on the multifaceted nature of energy transition and environmental impact.

```python
country = 'United States'
df_country = df[df['Entity'] == country]

fig, ax1 = plt.subplots(figsize=(12,6))

ax1.set_xlabel('Year')
ax1.set_ylabel('GDP per Capita (USD)', color='tab:blue')
ax1.plot(df_country['Year'], df_country['gdp_per_capita'], color='tab:blue', label='GDP per Capita')
ax2 = ax1.twinx()
ax2.set_ylabel('CO2 Emissions (kt)', color='tab:red')
ax2.plot(df_country['Year'], df_country['co2_emissions_kt'], color='tab:red', label='CO2 Emissions')

plt.title(f'GDP Per Capita vs. CO2 Emissions in {country} (2000-2020)')
plt.show()
```
